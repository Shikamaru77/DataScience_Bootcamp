```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris['data'], columns = iris['feature_names'])
df['species'] = [iris.target_names[n] for n in iris.target]
df
#here, iris.target = 0, 1 & 2 and target_names are just setosa, etc
```

## Univariate Analysis


```python
df_setosa = df.loc[df['species']== 'setosa']
df_virginica = df.loc[df['species']== 'virginica']
df_versicolor = df.loc[df['species']== 'versicolor']
df_setosa.head()
```

```python
iris['target_names']
```

```python
# performin univariate analysis by taking sepal_length as a feature to analyse for target
import matplotlib.pyplot as plt
plt.scatter(df_setosa['sepal length (cm)'], [0] * len(df_setosa), label='Setosa', marker='o')
plt.scatter(df_virginica['sepal length (cm)'], [1] * len(df_virginica), label='Virginica', marker='o')
plt.scatter(df_versicolor['sepal length (cm)'], [2] * len(df_versicolor), label='Versicolor', marker='o')

plt.xlabel('Sepal Length (cm)')
plt.yticks([0, 1, 2], ['Setosa', 'Virginica', 'Versicolor'])
plt.legend()
plt.show()
```
